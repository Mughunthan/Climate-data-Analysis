```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the CSV files into pandas dataframes
daily_data = pd.read_csv('daily_data.csv')
hourly_data = pd.read_csv('hourly_data.csv', dtype={'HourlyDryBulbTemperature': str})
monthly_data = pd.read_csv('monthly_data.csv')
three_hour_data = pd.read_csv('three_hour_data.csv')

# Data Cleaning and Preprocessing
# Convert date columns to datetime format
daily_data['DATE'] = pd.to_datetime(daily_data['DATE'])
hourly_data['DATE'] = pd.to_datetime(hourly_data['DATE'])
monthly_data['DATE'] = pd.to_datetime(monthly_data['DATE'])
three_hour_data['DATE'] = pd.to_datetime(three_hour_data['DATE'])

# Replace 'T' values with 0 for precipitation columns
daily_data['DailyPrecipitation'].replace('T', 0, inplace=True)
monthly_data['MonthlyTotalLiquidPrecipitation'].replace('T', 0, inplace=True)

# Convert precipitation columns to numeric
daily_data['DailyPrecipitation'] = pd.to_numeric(daily_data['DailyPrecipitation'], errors='coerce')
monthly_data['MonthlyTotalLiquidPrecipitation'] = pd.to_numeric(monthly_data['MonthlyTotalLiquidPrecipitation'], errors='coerce')

# Check column names in monthly_data
print(monthly_data.columns)

# Exploratory Data Analysis (EDA)
# Plotting temperature trends over time
plt.figure(figsize=(10, 6))
plt.plot(daily_data['DATE'], daily_data['DailyAverageDryBulbTemperature'], label='Daily Avg Temp')
plt.plot(hourly_data['DATE'], hourly_data['HourlyDryBulbTemperature'], label='Hourly Avg Temp')
plt.plot(monthly_data['DATE'], monthly_data['MonthlyMeanTemperature'], label='Monthly Avg Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (°F)')
plt.title('Temperature Trends Over Time')
plt.legend()
plt.show()

# Statistical Analysis
# Calculate mean, median, and standard deviation of precipitation
precip_mean = daily_data['DailyPrecipitation'].mean()
precip_median = daily_data['DailyPrecipitation'].median()
precip_std = daily_data['DailyPrecipitation'].std()
print(f'Precipitation Mean: {precip_mean}, Median: {precip_median}, Std Deviation: {precip_std}')

# Seasonal Analysis
# Plotting monthly precipitation
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_data['DATE'].dt.month, y=monthly_data['MonthlyTotalLiquidPrecipitation'])
plt.xlabel('Month')
plt.ylabel('Total Precipitation (inches)')
plt.title('Monthly Precipitation')
plt.show()

# Time Series Analysis
# Decompose time series data to analyze trends, seasonality, and residuals
decomposition = seasonal_decompose(daily_data['DailyAverageDryBulbTemperature'], model='additive', period=365)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(daily_data['DATE'], daily_data['DailyAverageDryBulbTemperature'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(daily_data['DATE'], trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(daily_data['DATE'], seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(daily_data['DATE'], residual, label='Residuals')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
